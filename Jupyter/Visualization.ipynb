```julia
using Makie, CairoMakie
```

```julia
fig = Figure(resolution = (1200,1200))
ax1 = Axis(fig[1,1])
ax2 = Axis(fig[1,2])
ax3 = Axis(fig[2,1:2])
```

```julia
x = collect(1:0.1:10)
y1 = sin.(x)
y2 = cos.(x).^2
nothing
```

```julia

scatter!(ax1,x,y1)

lines!(ax1,x,y1)
lines!(ax2,x,y2)
lines!(ax3,x,y2)
fig
```

```julia
tbl = (x = [1, 1, 1, 2, 2, 2, 3, 3, 3],
       height = 0.1:0.1:0.9,
       grp = [1, 2, 3, 1, 2, 3, 1, 2, 3],
       grp1 = [1, 2, 2, 1, 1, 2, 1, 1, 2],
       grp2 = [1, 1, 2, 1, 2, 1, 1, 2, 1]
       )

barplot(tbl.x, tbl.height,
        stack = tbl.grp,
        color = tbl.grp,
        axis = (xticks = (1:3, ["left", "middle", "right"]),
                title = "Stacked bars"),
        )
```

```julia
CairoMakie.activate!(type = "svg")

```

```julia
 # Figure関数を呼び出す。resolution = (縦の大きさ, 横の大きさ)の引数を取ることが出来る。
  # また、 backgroundcolor="色"を指定することで描写するFigure レイアウトの色を指定できる
  fig = Figure(resolution = (1200,1200), backgroundcolor="gray", figure_padding=1)
  # 使用するグラフの形状のレイアウトを決める
  # Axisを使用してFigureを分割する
  ax1 = Axis(fig[1,1], 
                titlesize = 30,
                ylabelsize = 30,
                xlabelsize = 30,
                title = "sin(x)",
                xlabel="x",
                ylabel="y"
                )
  ax2 = Axis(fig[1,2], 
  titlesize = 30,
  ylabelsize = 30,
  xlabelsize = 30,
  title = "cos(x)",
  xlabel="x",
  ylabel="y"
  )
  ax3 = Axis(fig[2,1:2], 
  titlesize = 30,
  ylabelsize = 30,
  xlabelsize = 30,
  title = "sin(x) + cos(x)",
  xlabel="x",
  ylabel="y"
  )
  x = collect(1:0.1:10)
  scatter!(ax1, x, sin.(x) )
  ax1.title = "sin(x)"
  ax1.xlabel="x"
  ax1.ylabel="y"
  lines!(ax2, x, cos.(x) )
  scatter!(ax3, x, sin.(x) + cos.(x) )
  save("../documents/Visualization/example1.pdf",fig)
```
